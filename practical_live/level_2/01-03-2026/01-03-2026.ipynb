## Data Visualisation

### Principles of Good visualisation
 - Clarity - easy to interpret
 - Proper Scale
 - It should be very simple - minimalism
 - Label Everything
 - Use color wisely
 - choose the correct chart for your data


#### Types of Visuals
 - Scatter plots ( shows how data set is scattered across x and y axis)
 - Line Charts (shows trend over time)
 - Bar Plots (shows how a categories are compared to each other)
 - stacked bar plot (shows how difference categories make up to 100%)
 - pie chart (shows categories that add up to 100%)
 - Box Plot/ Whisker plot (max, median, min, outliers and distribution)

## Cleaning and Feature Engineering

```python
df.info()
```

```python
### removing null customer id rows
df= df.dropna(subset=["Customer ID"])
df.info()
```

```python
### removing cancellation invoices (Invoices that starts with C) WE USE THE INVERT FUNCTION (~)

df = df[~df["Invoice"].astype(str).str.startswith("C")]
df.shape
```

```python
df.info()
```

```python
## removing zero and negative quantities

df = df[df["Quantity"]> 0]
df.shape
```

```python
## removing negative prices
df = df[df["Price"]> 0]
df.shape
```

```python
df["Revenue"] = df["Quantity"] * df["Price"]
df.info
```

```python
df.info()
```

```python
## convert invoice date to proper date
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
## creating another column as month and year from invoicedate
df["Month"] = df["InvoiceDate"].dt.to_period("M").astype(str)
```

```python
df.head()
```

```python
df["Year"] = df["InvoiceDate"].dt.year
df.info()
```

```python
df["Month_no"] = df["InvoiceDate"].dt.month
df.info()
```

```python
df.head()
```

 - create a canvas using width and height parameter, in the scale of inches
 - Decide on plot to use.
  - plot line ==> plt.plot() etc
 - Decide the title of plot
 - X axis label
 - Y axis label
   - rotate x label optional
 - fit layout, to prevent label clipping
   -  show grid optional
 - show the plot window or render the figure


```python
### plotting on a line plot

# create a canvas using width and height parameter, in the scale of inches
plt.figure(figsize=(12,6))
# Decide on plot to use.
plt.plot(monthly_revenue_groupby.index, monthly_revenue_groupby.values, marker = "o")
# plot line ==> plt.plot() etc
plt.title("Monthly revenue over time")
# Decide the title of plot
# X axis label
plt.xlabel("Month")
# Y axis label
plt.ylabel("Revenue")
# rotate x label optional
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)
#fit layout, to prevent label clipping
plt.tight_layout()
#show grid optional
plt.grid(True)
#show the plot window or render the figure
plt.show()
```

```python
## plotting multiple values in a single plot

font = {
"family": "serif",
"color": "blue",
"weight": "bold",
"size":8
}

# create a canvas using width and height parameter, in the scale of inches
plt.figure(figsize=(12,6), facecolor="white")

plt.style.use("classic")
# Decide on plot to use.
plt.plot(monthly_revenue_groupby.index, monthly_revenue_groupby.values, marker = "o", color= "green") ## changing the color of the line to green

for i, value in enumerate(monthly_revenue_groupby.values):
    plt.text(monthly_revenue_groupby.index[i], value + 10000, f'{int(value)}', ha="center", fontdict=font )


plt.plot(monthly_revenue_groupby.index, monthly_revenue_groupby.values+1000000, marker = "o") ## changing the color of the line to green

for i, value in enumerate(monthly_revenue_groupby.values):
    plt.text(monthly_revenue_groupby.index[i], (value+1000000) + 10000, f'{int(value+1000000)}', ha="center", fontdict=font )

# plot line ==> plt.plot() etc
plt.title("Monthly Revenue Over time")
# Decide the title of plot
# X axis label
plt.xlabel("Month")
# Y axis label
plt.ylabel("Revenue")
# rotate x label optional
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)
#fit layout, to prevent label clipping
#plt.tight_layout()
#show grid optional
plt.grid(True)
#show the plot window or render the figure
plt.show()
```

```python
#### show top 10 revenue generating countries

country_revenue = df.groupby("Country")["Revenue"].sum().sort_values(ascending=False).head(10)
country_revenue
```

```python
# create a canvas using width and height parameter, in the scale of inches
plt.figure(figsize=(12,6), facecolor="white")

plt.style.use("ggplot")
# Decide on plot to use.
plt.bar(country_revenue.index, country_revenue.values,color="blue") ## changing the color of the line to green

plt.title("Top 10 Countries by Revenue")
plt.xlabel("Country")
plt.ylabel("Total Revenue")
plt.xticks(rotation = 45)
plt.tight_layout()
#show grid optional
plt.grid(axis="x")

plt.show()
```
