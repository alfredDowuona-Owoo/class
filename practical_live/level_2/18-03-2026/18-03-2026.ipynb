## select

```python
for i in collection.find({}):
    print(i)

```

```python
# find all the category which belongs to smartphone
query = {
    "category" : "smartphone"
}
for i in collection.find(query):
    print(i)
```

```python
# find all the rows which has category key in it
query = {
    "category" : { "$exists" :True}
}
for i in collection.find(query):
    print(i)
```

```python
# get all the smartphone category where price is greater than 0

query = {
    "category" : "smartphone",
    "price" : {"$gt" : 0}
    # "price" >  0
}
for i in collection.find(query):
    print(i)
```

### insert

```python
new_product = {'name': 'iphone 17 pro', 
'category': 'smartphone', 
'price': 1600.0, 
'features': ['3-camera', 'display-retina'],
'manufacturer': {
    'name': 'foxconn tech', 
    'country': 'china'}
}

result = collection.insert_one(new_product)
print(f"Inserted document ID : {result.inserted_id}")
# 69bacea8e03d8aee07e916b8
```

### delete

```python
# 'user_name': 'monal'

query = {
    "user_name" : "monal",
}

result = collection.delete_many(query)
print(f"Deleted {result.deleted_count} documents.")
```

```python
from pandas.core.indexes.accessors import PeriodIndexFieldOps
from pymongo.common import _CaseInsensitiveDictionary
# search all the names which has pixel word in them
query = {
    "name" : {"$regex" : "pixel", "$options" : "i"}
}

# i - > case-_CaseInsensitiveDictionary
# # Pixel
# # pixel
# # PIXEL
# # piXEL

for i in collection.find(query):
    print(i)
```

### update

```python
query_filter = {
    "name" : {"$regex" : "17", "$options" : "i"}
}

update_operation = {
    "$set" : {
        "category" : "smartphone_latest"
    }
}

result = collection.update_many(query_filter, update_operation)

print(f"Matched {result.matched_count} documents and modified {result.modified_count} documents.")
```
