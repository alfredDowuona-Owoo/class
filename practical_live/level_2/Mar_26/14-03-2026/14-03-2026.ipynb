```python
import sqlite3

#step 1
conn = sqlite3.connect('student_1.db')

```

```python
#step 2
cursor = conn.cursor()
cursor.execute('PRAGMA foreign_keys = ON;') ## OPTIONAL ONLY FOR SQLITE
```

Create a table in sql 

```python
cursor.execute('''
    CREATE TABLE Ultimate_Date_Science (
        student_id INT PRIMARY KEY NOT NULL,
        name TEXT NOT NULL,
        batch_id INT NOT NULL    
               
               )
''')
```

```python
student_vals = [(1, "Alfred Owoo", 2), (2, "Nii Nortey", 2), (3, "Eileen", 2)]
```

```python
## Insert values in tuples
for student_val in student_vals:
    cursor.execute('''
        INSERT INTO Ultimate_Date_Science( student_id, name, batch_id)
        VALUES(?,?,?)
        ''',student_val)
```

```python
## save  --> push the data from memory to database
## saves the state of database
conn.commit()
```

```python
conn.close()
```
