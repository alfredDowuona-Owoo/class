## Series indexing and & accessing

```python
## 1. Positional indexing

print(s5.values[0])
print(s5.values[3])
```

#### 1. iloc positional based indexing

```python
## iloc gets rows (and / or columns ) at integer locations

print(s5.iloc[0])
```

```python
## iloc gets rows (and / or columns ) at integer locations

print(s5.iloc[1:3])
```

#### 2. label based indexing

#####  - loc gets rows (and/ or column) with particular labels

```python
print(s5["c"])
```

```python
print(s5["a"])
```

```python
print(s5["e"])
```

```python

print(s5[["b","d"]])
```

```python
## .loc[] uses the label/ index name not position
print(s5.loc[["b","e"]])
```

# Working with real data set

```python

titanic_dataset_github_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(titanic_dataset_github_url)

df.head()
```

```python
df["Age"]>70 ## only boolean index of passengers ages greater than 70
```

```python
## filtering the dataset
df[df["Age"]> 70] ### Gives us the actual values
```

```python
older_passengers = df[df["Age"]> 70] 
older_passengers
```

```python
## females who survived

## sex = female and survived = 1

(df["Sex"]== "female") & (df["Survived"] == 1 )  ### will just return either true or false
```

```python
## females who survived

## sex = female and survived = 1

df[(df["Sex"]== "female") & (df["Survived"] == 1 )]
```

```python
### get name and age of females who survived
female_survivors_subset = df[(df["Sex"]== "female") & (df["Survived"] == 1 )][["Name", "Age"]]
female_survivors_subset
```

```python
### get name and age of females who survived using loc
## loc requires the labels 
## df.loc[label, ["Name", "Age"]]
df.loc[(df["Sex"]== "female") & (df["Survived"] == 1 ), ["Name", "Age"]]
```
