### Array reshaping and resizing in numpy
1. reshape()
- Total number of elements remains the same
- reshape returns a new view if possible
- use -1 to let numpy infer one dimension

```python
arr = np.array([1,2,3,4,5,6])
print(arr.shape)
```

```python
reshape_1 = arr.reshape(3,2)
print(reshape_1)
```

```python
reshape_2 = arr.reshape(1,6)
print(reshape_2)
print(reshape_2.ndim)
```

```python
arr_w = np.array([1,2,3,4,5,6,10,200,789])
print(arr_w.shape)
```

```python
reshape_3 = arr_w.reshape(3,-1)
print(reshape_3)
print(reshape_3.ndim)
```

```python
reshape_3 = arr_w.reshape(4,-1)
print(reshape_3)
print(reshape_3.ndim)
```
