### sort, search and Counting functions

```python
arr = np.array([6,3,8,5,9,7,9,11,22,55,11,6,8,552,0])
print("sorted", np.sort(arr))
print("Arg Sort indices", np.argsort(arr)) ## sort by the index 
```

```python
print("insert 4 position in ", np.searchsorted(np.sort(arr), 4)) ## insert 4 in position 2
```

```python
### finding how many unique values

print("Unique element", np.unique(arr))

### finding the count of each elements

print("count of each element", np.bincount(arr)) ## start from min ==0 max 8
```

```python
arr4 = np.array([6,3,8,5,9,7,9,11,2,5,11,6,8,12,0,17,11,13,8,6,3])

print("insert 4 position in ", np.searchsorted(np.sort(arr4), 4)) ## insert 4 in position 2
### finding how many unique values

print("Unique element", np.unique(arr4))

### finding the count of each elements

print("count of each element", np.bincount(arr4)) ## start from min ==0 max 8
```

### Numpy char and strings

```python
## np.char.<function_name>()
```

```python
## np.add()

a= np.array(["hello", "world"])
b= np.array(["_user", "_123456"])
print(np.char.add(a,b))
```

```python
print(np.char.multiply(a,b))
```

```python
## np.char (lower, upper, capitalize)

a= np.array(["hello", "world"])
print(np.char.upper(a))
print(np.char.lower(a))
print((np.char.capitalize(a)))
```

```python
## np.char.center()

print(np.char.center(a,11,fillchar="*"))
```

```python
def password(value):
    #print(value[-1:-4:-1])
    p=np.array([value[-4:]])
    print(p)
    print(np.char.center(p,30,fillchar="*"))
```
