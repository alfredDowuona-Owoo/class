# Commodity Prices Dataset Analysis

Kaggle Dataset link : https://www.kaggle.com/datasets/debashish311601/commodity-prices

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("data/commodity_futures.csv")
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values("Date").reset_index(drop=True)
df.head()
```

```python
# Create Year & Month for seasonal/group analysis
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df.head()
```

```python
# Quick check of missing values
print(df.isna().sum())
```

```python
# Forward fill missing values
df_ffill = df.ffill()
```

```python
# Quick check of missing values
print(df_ffill.isna().sum())
```

```python
df_clean = df.ffill().bfill()
```

```python
# Check again
print(df_clean.isna().sum())
```

### Descriptive stats

```python
# Price Trends of Commodities (Descriptive)
## Metals
metals = ['ALUMINIUM', 'NICKEL', 'ZINC', 'COPPER', 'GOLD', 'SILVER']
plt.figure(figsize=(14,6))
for col in metals:
    plt.plot(df_clean['Date'], df_clean[col], label=col)
plt.title("Commodity Price Trends (2000–2024)")
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
# Price Trends of Commodities (Descriptive)
## Metals without Nickel
metals = ['ALUMINIUM',  'ZINC', 'COPPER', 'GOLD', 'SILVER']
plt.figure(figsize=(14,6))
for col in metals:
    plt.plot(df_clean['Date'], df_clean[col], label=col)
plt.title("Commodity Price Trends (2000–2024) (without Nickel)")
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
# Price Trends of Commodities (Descriptive)
## Metals just copper and silver
metals = ['COPPER', 'SILVER']
plt.figure(figsize=(14,6))
for col in metals:
    plt.plot(df_clean['Date'], df_clean[col], label=col)
plt.title("Commodity Price Trends (2000–2024) (Only Copper and Silver)")
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend()
plt.show()
```
