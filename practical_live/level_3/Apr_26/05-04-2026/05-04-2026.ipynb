## Quartiles

```python
import numpy as np
import pandas as pd
```

```python
student_scores1 = [10,12,45,34,20,33,35,40,55,60,61,62,68,70,44,48,53,90,98]
student_scores1.sort()
```

```python
def detect_outliers(data, k=1.5):
    # sort data
    data.sort()

    arr = np.array(data, dtype=float)

    #Q1 & Q3
    Q1=np.percentile(arr, 25, method='linear')
    Q3=np.percentile(arr, 75, method='linear')

    #IQR
    IQR = Q3-Q1
    
    # LOWER FENCE AND UPPER FENCE
    lower = Q1 - k * IQR
    upper = Q3 - k * IQR

    mask = (arr>=lower) & ( arr<=upper ) ## mask to remove outliers
    outliers_mask = ~mask # to extract outliers

    return{
        "Q1": float(Q1),
        "Q3": float(Q3),
        "IQR": float(IQR),
        "LOWER_FENCE": float(lower),
        "UPPER_FENCE": float(upper),
        "Outliers": arr[outliers_mask].tolist(),
        "Clean_Data": arr[mask].tolist()

    }


# lower = 20
# upper = 90
# (value > 20) &(value< 90)
# False, False, True, ..... False, False -->mask 
```

```python
student_scores = [10,12,45,34,20,33,35,40,55,44,48,53,90,98]
print(detect_outliers(student_scores))
```

## kurtosis

```python
from scipy.stats import kurtosis
```

```python
data  = [10,20,30,40,90]
mean = np.mean(data)
std = np.std(data)

skew = np.mean((data-mean)**3)/(std**3)
kur = (np.mean(data - np.mean(data))**4 / (np.std(data)**4))-3

print("skew:- ", skew)
print("kurtosis_old: - ", kur)
print("kurtosis:-", kurtosis(data))
```
