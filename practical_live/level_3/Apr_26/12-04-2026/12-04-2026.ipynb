# 1. Bernoulli Distribution
- is discrete with only two values that is either pass/fail, yes/no, 0/1, head/tail

- Mathematical Properties
- For a Bernoulli random variable $X \sim \text{Bernoulli}(p)$:
---
> Probability Mass Function (PMF)
- The probability of a specific outcome $k$ is given by:
$$f(k;p) = p^k (1-p)^{1-k} \quad \text{for } k \in \{0,1\}$$

- Mean (Expected Value)The average outcome over many trials:$$E[X] = p$$
- VarianceThe measure of spread or uncertainty in the trial:$$Var(X) = p(1 - p)$$
---
> Practical Applications
- In data science and Boolean logic, this distribution is used to model binary classifications:

- A/B Testing: Whether a user clicked a button (1) or did not (0).

- Quality Control: Whether a single item is defective (1) or functional (0).

- Medical Trials: Whether a treatment was effective (1) or ineffective (0).



```python

from scipy.stats import bernoulli
p = 0.6 ## for success
```

```python
print(bernoulli.pmf(1,p) )
```

# 2. Binomial distribution
 - model n bernoulli trials
 - A Binomial distribution is the result of multiple independent Bernoulli trials added together (e.g., flipping a coin 10 times and counting the total number of heads).
> 1. Core ConceptFor a random variable $X$ to follow a Binomial distribution, four specific conditions must be met (often remembered by the acronym BINS):
* Binary: Each trial has only two possible outcomes (Success or Failure).
* Independent: The outcome of one trial does not affect the outcome of another.
* Number: There is a fixed number of trials ($n$).
* Success: The probability of success ($p$) is the same for every trial.
* Example: Flipping a fair coin 10 times ($n=10$, $p=0.5$) and counting how many times it lands on "Heads."
---
> Mathematical Properties
 - A Binomial random variable is denoted as $X \sim B(n, p)$.
 - Probability Mass Function (PMF)The probability of getting exactly $k$ successes in $n$ trials is:$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$Where:$\binom{n}{k}$ is the binomial coefficient, calculated as $\frac{n!}{k!(n-k)!}$ ("$n$ choose $k$").
 - $p^k$ is the probability of $k$ successes.
 - $(1-p)^{n-k}$ is the probability of the remaining failures.
 - Mean (Expected Value)
   -  The average number of successes you would expect over many repetitions:$$E[X] = np$$
 - Variance
   -  The measure of spread or risk around the mean:$$Var(X) = np(1 - p)$$

# Poisson Distribution
 > The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space, provided these events occur with a known constant mean rate and independently of the time since the last event.
  - count of rare events occuring at constant rate ($\lambda$) in a fixed internal.
---
Core Concept
> A random variable $X$ follows a Poisson distribution if it meets these conditions:
  - Independence: Events occur independently. The occurrence of one event does not affect the probability of another.  
  - Constant Rate: The average number of events per unit of time/space ($\lambda$) is constant.
  - Singularity: Two events cannot happen at the exact same instant.
  - Example: The number of emails received per hour, or the number of network failures per month.
---
> A Poisson random variable is denoted as $X \sim \text{Poisson}(\lambda)$.
 - Probability Mass Function (PMF)
   - The probability of observing exactly $k$ events in an interval is:$$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$$
Where:
 - $\lambda$ (Lambda) is the average number of events in the given interval.
 - $e$ is Euler's number ($\approx 2.71828$).
 - $k$ is the number of occurrences ($k = 0, 1, 2, \dots$).
 - $k!$ is the factorial of $k$.
- Mean (Expected Value)$$E[X] = \lambda$$
- Variance
  - A unique property of the Poisson distribution is that its mean and variance are equal:$$Var(X) = \lambda$$




# Exponential Distribution

> The Exponential distribution is a continuous probability distribution that models the time between events in a Poisson process. While the Poisson distribution counts the number of events in a fixed interval, the Exponential distribution measures the waiting time until the next event occurs.
---
>  Core Concept
-  key characteristic of the Exponential distribution is that it is memoryless. This means the probability of an event occurring in the next 10 minutes is the same regardless of how much time has already passed.
-  Continuous: Unlike the Binomial or Poisson distributions, which deal with discrete counts ($0, 1, 2...$), the Exponential distribution deals with continuous data (time, distance, etc.).
-  Relationship to $\lambda$: It is defined by the rate parameter $\lambda$ (the same lambda used in Poisson), representing the average number of events per unit of time.
-  Example: If a database receives an average of 10 queries per minute ($\lambda = 10$), the Exponential distribution models the specific time gap (in seconds or milliseconds) between each query.

An Exponential random variable is denoted as $X \sim \text{Exp}(\lambda)$.Probability Density Function (PDF)For $x \ge 0$:$$f(x; \lambda) = \lambda e^{-\lambda x}$$Cumulative Distribution Function (CDF)The probability that an event occurs within $x$ units of time:$$F(x; \lambda) = 1 - e^{-\lambda x}$$Mean (Expected Value)The average waiting time between events:$$E[X] = \frac{1}{\lambda}$$Variance$$Var(X) = \frac{1}{\lambda^2}$$

# Uniform Distribution

> The Uniform distribution is a probability distribution where every possible outcome has an equal probability of occurring within a specific range. It is the mathematical representation of "perfect fairness."
 - every value in the interval [a,b] is equally likely.
 - the pdf is a flat rectangle, constant height

Uniform distributions are divided into two types: Discrete (for specific, separate values) and Continuous (for any value within a range).

> 1. Discrete Uniform Distribution
This occurs when there are a finite number of outcomes, and each is equally likely.
- Definition: If a random variable $X$ can take $n$ possible values, the probability of each value is $1/n$.
- Example: Rolling a fair six-sided die. Each number ($1, 2, 3, 4, 5, 6$) has a probability of exactly $1/6$.
- Mathematical Properties:
     - Mean: $E[X] = \frac{a + b}{2}$
     - Probability Mass Function (PMF): $P(X=k) = \frac{1}{n}$2. 

> 2. Continuous Uniform Distribution
This occurs when $X$ can take any real number value between a minimum ($a$) and a maximum ($b$).
- Definition: The probability density is constant across the interval $[a, b]$ and zero everywhere else.
- Example: If a bus arrives exactly every 15 minutes, and you arrive at a random time, your waiting time is uniformly distributed between 0 and 15 minutes.
- Mathematical Properties:
- For $X \sim U(a, b)$:
    - Probability Density Function (PDF):$$f(x) = \begin{cases} \frac{1}{b - a} & \text{for } a \le x \le b \\ 0 & \text{otherwise} \end{cases}$$
    - Mean (Expected Value):$$E[X] = \frac{a + b}{2}$$
    - Variance:$$Var(X) = \frac{(b - a)^2}{12}$$


> Characteristics
 - The "Rectangle" Distribution: The graph of a continuous uniform distribution is a rectangle because the height (probability density) is constant. The area of this rectangle must always equal 1.

 - Maximum Entropy: Among all distributions supported on a specific interval, the uniform distribution has the highest uncertainty (entropy), as it doesn't favor any particular value.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 samples between 0 and 10
data = np.random.uniform(0, 10, 1000)

plt.hist(data, bins=20, density=True, alpha=0.7,edgecolor='black', color='skyblue')
plt.title("Continuous Uniform Distribution")
plt.show()
```
