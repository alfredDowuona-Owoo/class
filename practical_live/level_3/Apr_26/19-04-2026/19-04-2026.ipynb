## confidence interval


```python
import numpy as np
from scipy import stats
```

```python
sample_mean = 31
pop_sd = 4
n = 36
confidence= 0.95
```

```python
#SE
se = pop_sd/np.sqrt(n)
se
```

```python
## calculate the interval
lower, upper = stats.norm.interval(confidence, loc=sample_mean, scale=se)
```

```python
print(f"95% CI --- Lower bound : {lower:.4f}, upper bound : {upper:.4f}")
```

## t-test

```python
bulb_data = [980,960,1020,940,950]
claimed_mean = 1000

t_Stat, p_val = stats.ttest_1samp(bulb_data, claimed_mean)

print(f"t-statistics: {t_Stat:.4f}")
print(f"p_value: {p_val:.4f}")
```
