## (1/n) vs (1/(n-1))

```python
np.random.seed(88)
```

```python
population = np.random.normal(loc=50, scale=10, size=100000)

#normal  = normal distribution
# loc = mean
# scale  = SD = 10
```

```python
population.shape
```

```python
population[1:10]
```

```python
True_variance = np.var(population)
print("True_variance: -", True_variance)
```

```python
n_samples = 1000
sample_Size = 5

var_using_n = []
var_using_n_minus_1 = []
var_using_n_minus_2 = []
```

```python
for _ in range(n_samples):
    sample = np.random.choice(population, size=sample_Size, replace=False)

    var_n = np.var(sample)
    var_n1 = np.var(sample, ddof=1)
    var_n2 = np.var(sample, ddof=2)

    var_using_n.append(var_n)
    var_using_n_minus_1.append(var_n1)
    var_using_n_minus_2.append(var_n2)
```

```python
avg_Var_n = np.mean(var_using_n)
avg_Var_n1 = np.mean(var_using_n_minus_1)
avg_Var_n2 = np.mean(var_using_n_minus_2)
```

```python
print("Average with n:-", avg_Var_n)
print("Average with n-1:-", avg_Var_n1)
print("Average with n-2:-", avg_Var_n2)
```

- for variance for population we divide by n
- for variance of sample we divide by n-1
- dividing by n-1 instead of n, inflates the results just enough to cancel out the underestimate called BESSEL CORRECTION
