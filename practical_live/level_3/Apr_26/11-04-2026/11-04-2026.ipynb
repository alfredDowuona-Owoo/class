## Covariance and Correlation

```python
import numpy as np
import pandas as pd
x= [1,2,3]
y = [30,40,50]
```

```python
print(np.corrcoef(x,y))
```

```python
# Create the DataFrame
df = pd.DataFrame({
    'x': x,
    'y': y
})
```

```python
# Display the DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlation matrix
corr = df.corr()

# Create heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap.png')
```

## Correlation using titanic dataset

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])
```

```python
numeric_df.columns
```

```python
corr_matrix = numeric_df.corr()
```

```python
corr_matrix
```

```python
# Calculate correlation matrix
plt.figure(figsize=(10,6))

# Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths= 0.5, fmt=".2f")
plt.title('Correlation Heatmap of Titanic Dataset Numeric Features')
plt.show()
#plt.savefig('correlation_heatmap.png')
```

```python
bins = [150,160,165,170,175]
pdf_Values = [0.01,0.00,0.03,0.06,0.06,0.04]
```

```python
plt.bar(
    x=bins, 
    height=pdf_Values,  # Use 'height' for your y-axis data
    width=5,            # Width of the bars
    align="center", 
    alpha=0.5,           # If you meant 'float' to be transparency, use 'alpha'
    edgecolor='black',  # This sets the border color
    linewidth=1.5       # This sets the thickness of the border
)

plt.title('PDF of heights')
plt.xlabel('Height')
plt.ylabel('Probability Density')
plt.show()
```
