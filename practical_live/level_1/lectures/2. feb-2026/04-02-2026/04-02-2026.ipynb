## OOPs Intro

### procedural Programming

- Simple bank account system

```python
user_name = "Alice"
user_balance = 100.0

def withdraw(balance, amount):
    if amount > balance:
        print("Insufficient funds.")
        return balance
    else:
        balance -= amount
        print(f"Withdrew ${amount:.2f}. New balance: ${balance:.2f}")
        return balance
```

```python
user_balance = withdraw(user_balance, 30.0)
print(f"{user_name}'s current balance: ${user_balance:.2f}")
```

```python
user1_name = "Firoz"
user1_balance = 10000

user2_name = "Abhinav"
user2_balance = 20000

user1_balance = withdraw(user1_balance, 2000)
user2_balance = withdraw(user2_balance, 20000)

print(f"{user1_name}'s current balance: ${user1_balance:.2f}")
print(f"{user2_name}'s current balance: ${user2_balance:.2f}")
```

#### user defined data structure

```python
class OurBankAccount:
    def __init__(self, name, balance):
        self.name=name
        self.balance=balance

    def withdraw(self, amount):
        if amount >self.balance:
            print("insufficient Balance")
        else:
            self.balance-=amount
            print(self.name, "withdrew", amount, "new balance is:", self.balance)
```

```python
user_1 = OurBankAccount("Monal", 10000)
```

```python
user_1.withdraw(6000)
```

```python
print(type(user_1))
```

```python
user_2 = OurBankAccount("deb", 10000)
```

```python
user_2.withdraw(10000)
```

- function inside a class is called a method
- functions outside a class is called a function
- to call a property of a class object we need to use the dot "." operator

```python
class Robot:
    def __init__(self,  customer_Name):
        print("hello welcome to Nii Robot")
        self.name= "chitti"
        self.ram_in_gb = 10
        self.version = "0.0.1"
        self.customer_name = customer_Name
    
    def config(self):
        print("Ram : ", self.ram_in_gb)
        print("version : ", self.version)
        print("Robot Name : ", self.name)


    def customer_details(self):
        print("Customer Name: ", self.customer_name) 
```

```python
### when u start ur laptop

r1=Robot("Alfred")
```

```python
r1.config()
```

```python
r1.customer_details()
```
