#### Encapsulation

- Public : { "syntax" : "name", "access": "anywhere"}
- Protected : { "syntax" : "_name", "access": "class + child access"}
- Private : { "syntax" : "__name", "access": "only inside class"}

public, protected and private assumes the developer is responsible and consent

#### Inheritance

- Inheritance allows a class to reuse the properties of another class
- types of Inheritance
 - single inheritance  - class b inherit from class a
 - multilevel inheritance  - class c inherits from class b and class b inherits from class a
 - multiple inheritance - class c inherits from both class a and class b

```python
# single Inheritance

class Device:
    def power_on(self):
        print("device on")

class TV(Device): ### we inherit the device
    def show_channels(self):
        print("channels 1,2,3,4,5")
```

```python
tv_1 = TV()
```

```python
tv_1.show_channels()
```

```python
tv_1.power_on()
```

```python
## pass  -  python read the code and ignores the line itself
```

```python
# multilevel Inheritance

class Device:
    def a (Self):
        print("abc")

class TV(Device):
    def a2(self):
        print("abc")
    
class smartbulb(TV):
    def a3(self):
        print("abc")

```

```python
sbml=smartbulb()
```

```python
sbml.a3()
sbml.a2()
sbml.a()
```

```python
### multiple inheritance

class Camera:
    def click_photo(self):
        print("photo")


class Phone:
    def call(self):
        print("call")

class Smartphone(Camera, Phone):
    def video_call(self):
        print("video call")
    
```

```python
smp= Smartphone()
```

```python
smp.call()
smp.click_photo()
smp.video_call()
```

```python
class A:
    def speak(self):
        print("parent speak")

class B(A):
    def speak(self):
        print("CHILD speak")
```

```python
## the use of Super() to call the main class or reference it.

class A:
    def speak(self):
        print("parent speak")

class B(A):
    def speak(self):
        super().speak() ## a().speak()
        print("CHILD speak")
```

```python
ob = B()
```

```python
ob.speak()
```

```python
## the use of Super() to call the main class or reference it.

class A:
    def __init__(self):
        print(" I am inside A")
    def speak(self):
        print("parent speak")

class B(A):
    def __init__(self):
        super().__init__()
        super().speak()


    def speak(self):
        #super().speak() ## a().speak()
        print("CHILD speak")
```

```python
class Motherboard:
    def __init__(self, ram, cpu_speed, cpu_type):
        self.ram=ram
        self.cpu_speed=cpu_speed
        self.cpu_type = cpu_type

    def start(self):
        print("power flow started")

    def stop(self):
        print("power flow stopped")
        
```

```python
class Computer(Motherboard):
    def __init__(self, ram, cpu_speed, cpu_type):
        super().__init__(ram, cpu_speed, cpu_type) ## super() inherit or call the parent directly which is motherboard
        #
        self.os = self.__check_os_and_set_os_type()

    def __check_os_and_set_os_type(self): ## PRIVATE METHOD
        if self.cpu_type.lower() == "mac-chip":
            return "mac"
        else:
            return "Other"
        
    def check_my_os(self): # PUBLIC METHOD
        print("user has :", self.os, "OS")
```

```python
Cmp1 = Computer("12gb", "4.2ghz", "mac_chip")
```

```python
Cmp1.check_my_os()
```

```python
Cmp1.start()
```

```python
Cmp1.stop()
```

```python
Cmp1.cpu_speed
```

```python
Cmp1.cpu_type
```

```python
### Method resolution order

# used to find next method in the inheritance chain

class A:
    def __init__(self):
        print("Class A")

    def a(self):
        print("a")

class B(A):
    def __init__(self):
        print("Class B")

    def b(self):
        print("b")

class C(A):
    def __init__(self):
        print("Class C")

    def c(self):
        print("c")

class D(B,C):
    def __init__(self):
        super().__init__() # MRO happens here - used to find next method in the inheritance chain
        print("Class D")

    def d(self):
        print("d")
```

```python
d_obj = D()
```

```python
print(D.__mro__)
```

```python
### Method resolution order

# used to find next method in the inheritance chain

class A:
    def __init__(self):
        print("Class A")

    def a(self):
        print("a")

class B(A):
    def __init__(self):
        super().__init__()
        print("Class B")

    def b(self):
        print("b")

class C(A):
    def __init__(self):
        print("Class C")

    def c(self):
        print("c")

class D(B,C):
    def __init__(self):
        super().__init__() # MRO happens here - used to find next method in the inheritance chain
        print("Class D")

    def d(self):
        print("d")
```

### ABSTRACTION (Contract)


```python
from abc import ABC, abstractmethod
```

```python
## abstract class

class Payment(ABC): ## Requirement number 1
    @abstractmethod  ## Requirement number 2
    def pay(self, amount, currency): ## arguement matching not mandatory
        pass

    @abstractmethod ## Requirement number 2
    def refund(self):  ## arguement matching not mandatory
        pass

## contract called payment has 2 methods in the contract which has all the bank needs to implement 
# first is pay
# second is refund
```

```python
class IPLDBank(Payment):

    def pay(self):
        print("PAY")

    def refund(self):
        print("REFUND")


    def new_user(self):
        print("HELLO")
```

```python
obb = IPLDBank()
```

```python
obb.pay()
```

```python
obb.refund()
```

```python
obb.new_user()
```
