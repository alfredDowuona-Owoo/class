- Dunder Methods egs
- 1. __init__ ==> helps us initialize the object variables (attributes).
- 2. __str__ ==> description of the class and intended for users.
- 3. __repr__ ==> description but intended for developers. can be used for logging




```python
class Car:
    def __init__(self, manuf, car_name):
        self.manuf= manuf
        self.car_name = car_name

    def __str__(self):
        return f"manufacturer name : {self.manuf} and we produce car"
```

```python
class Car:
    def __init__(self, manuf, car_name):
        self.manuf= manuf
        self.car_name = car_name

    def __str__(self):
        return f"manufacturer name : {self.manuf} and we produce car"
    
    def __repr__(self):  ## official representation of an object can be used for logging
        return f"manufacturer name : {self.manuf} and we produce car this object contains car model : {self.car_name}"
```

```python
car_2 = Car("Toyota", "vits")
```

```python
str(car_2)
```

```python
repr(car_2)
```

```python
car_2 ## with REPR we get the description now
```

```python
print(f"{car_1!s}") ## __str__
```

```python
print(f"{car_2!r}") ## __repr__ = official representation 
```

```python
class Account:
    def __init__ (self, balance):
        self.__balance = balance


    def balance_Set(self, value):
        print("changing balance")
        self.__balance = value

    def balance_get(self, value):
        print("changing balance")
        self.__balance = value

    def balance_det(self, value):
        print("changing balance")
        self.__balance = value

    balance = property(fset=balance_Set) ## apply setter to a method balance_set
    balance = property(fset=balance_get) ## apply get to a method balance_set
    balance = property(fset=balance_det) ## apply det to a method balance_set
    
```

```python
class Account:
    def __init__ (self, balance):
        self.__balance = balance

 #   def set_balance(self, value):
 #       self.__balance = value

 #   def get_balance(self):
 #       return self.__balance ### old method


### new methods
### get value, set value, delete value

### getter
    @property ## we create a getter using properter
    def balance(self):
        print("checking the balance")
        return self.__balance
    

    ##setter 
    @balance.setter
    def balance(self, value):
        if value < 0:
            print("Security Error . Balance cannot be negative")
        elif value > 1000000:
            print("Security Error. Large transaction require approval")
        else:
            print("changing balance")
            self.__balance = value
   
   ## Delete 
    @balance.deleter
    def balance(self):
        print(f"ERROR: You cannot delete a balance record")
```

### Filter

 - filter (function, iterable)
 - function that takes one item and returns a boolean

```python
list_c = [1,2,3,5,6,7,8, 10,11,13,16]

def is_even(num):
    # 2 --> 2%2==0 - > true
    # 3 --> 3%2==0 - > false
    return num % 2 == 0
```

```python
list(filter(is_even, list_c))
```

### List and Dictionary Comprehension

```python
num = list(range(1,21))
print(num)
```

```python
def even_number(l):
    final_even_list = []
    for i in l:
        if i%2 == 0 :
            final_even_list.append(i)
    return final_even_list
```
