```python
## we would have normally written it this way

class TV:
    def start_tv(self):
        print("TV")

class Microwave:
    def start_mv(self):
        print("mv")

class Fan:
    def start_fan(self):
        print("FAN")
```

```python
tv= TV()
tv.start_tv()

mv= Microwave()
mv.start_mv()

fans=Fan()
fans.start_fan()
```

```python
## with polymorphism concept

class TV:
    def start(self):
        print("TV")

class Microwave:
    def start(self):
        print("mv")

class Fan:
    def start(self):
        print("FAN")
```

```python
electronic_appliance = [TV(), Microwave(), Fan()]
```

```python
for e in electronic_appliance:
    e.start()
```

```python
class TV:
    def start(self):
        print("TV")

class Microwave:
    def start(self):
        print("mv")

class Fan:
    def start(self):
        print("FAN")

class Airfrier:
    def start(self):
        print("AF")
```

```python
from abc import ABC, abstractmethod
```

```python
class ElectronicContract(ABC):


    @abstractmethod
    def start(self):
        pass
```

```python
class TV(ElectronicContract):
    def start(self):
        print("TV")

class Microwave(ElectronicContract):
    def start(self):
        print("mv")

class Fan(ElectronicContract):
    def start(self):
        print("FAN")

class Airfrier(ElectronicContract):
    def start(self):
        print("AF")

    def start(self):
        print("AF")
```

```python
electronic_appliance = [TV(), Microwave(), Fan(), Airfrier()]
```

```python
for e in electronic_appliance:
    e.start()
```
