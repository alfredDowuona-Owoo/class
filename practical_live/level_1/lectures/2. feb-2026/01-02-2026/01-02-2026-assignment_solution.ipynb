```python
def mark_attendance():
    attendance_set = set()

    while True:
        names = input("Enter your name as a student here :")

        if names.upper() == "STOP":
            break

        if names:
            attendance_set.add(names.title())

    print(f"\n ----attendance Report -----")
    print(f"Total Unique Students: {len(attendance_set)}")
    print(f"Attendees : {", ".join(attendance_set)}")
mark_attendance()
```

```python
def split_bills(people, tax= 0.05, tip = 0.10):
    items_prices = []

    while True:

        sub_total_bill = input("Enter the price list of Items:")
        if sub_total_bill.upper() == "STOP":
            break
        items_prices.append(float(sub_total_bill))

    sum_bills = sum(items_prices)
    totals = sum_bills * (1+ tax + tip)
    share = round((totals/people), 2)
    
    
    print(f"The bills are {items_prices}")
    print(f"Total bill was : {sum_bills}")
    print(f"Total bill with tax and tip is : {totals}")
    print(f"Bill will be share by {people}")
    print(f"Each person will pay GHS{share}")

    return share


split_bills(5)
```

```python
def split_bills(people, tax=0.05, tip=0.10):
    
    items_prices = []

    while True:
        price = input("Enter item price (type STOP to finish): ")

        if price.upper() == "STOP":
            break

        items_prices.append(float(price))

    sub_total = sum(items_prices)

    total = sub_total * (1 + tax + tip)

    share = round(total / people, 2)

    print(f"Items: {items_prices}")
    print(f"Total bill: {sub_total}")
    print(f"Each person pays: GHS {share}")

    return share

split_bills(5)
```

```python
def apply_discount(total):
    if total > 500:
        discount = 0.2
    elif total > 200:
        discount = 0.1
    else:
        discount = 0.0
    discount_amount = total * discount
    price_to_pay = total - discount_amount
    return price_to_pay

apply_discount(100)
```

```python
def validate_pin():
    correct_pin = "1234"
    attempt = 0

    while attempt< 3:
        user_pin = input(f"Enter Your Pin {3- attempt} attempts left:")
        if user_pin == correct_pin:
            print(f"Access granted, Successful login Thank you")
            return
        attempt+=1
        if attempt == 3:
            print("Incorrect Pin inputted. 3 times")
    print("🚫 Error Pin attempt exceeded, Card Blocked")

validate_pin()
```

```python
def create_invite(event_name, *guest_names, **details):
    print(f"\n --- {event_name.upper()}----")

    print(f"Guest List :")
    for guest in guest_names:
        print(f" ✅ {guest}")

    print("Event Details are: ")
    for key, value in details.items():
        formatted_key = key.replace("_", " ").title()
        print(f"{formatted_key}: {value}")

create_invite("Python Workshop", "Alice", "Bob", Date="Jan 24, 2026", Location="Room 404", Time="10 AM")
```
