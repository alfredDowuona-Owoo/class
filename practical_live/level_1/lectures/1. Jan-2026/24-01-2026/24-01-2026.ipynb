### Non primitive data types
#### List

- when we append something to the list it maintains the order to the append
- accepts any data type and duplicates are allowed
- methods
  - append() add an element or value at the end
  - pop() remove by index
  - index [] tells us the position of an element /retrieve values by indexing and starts with 0 and from the end it starts from -1
  - len() - string counts characters/ list counts number of items or elements
  - remove() - accepts values if we do not know the index. removes first occurrence
  - insert() -  insert elements between values

```python
empty_list = []  # Creating an empty list
print(type(empty_list))  # Checking the type of empty_list

### or 
empty_list1 = list()  # Creating an empty list using the list() constructor
print(type(empty_list1))  # Checking the type of empty_list1
```

```python
l = ["python", "java", "c++", "ruby", "javascript", 10, 25.5, True, None, [1, 2, 3]]    
```

```python
print(l[3])  # Accessing the fourth element (index 3)
print(l[-1])  # Accessing the last element (index -1)
```

```python
len("data science")
```

```python
len(l)  # Getting the length of the list
```

```python
l[len(l) - 1]  # Accessing the second last element using len()
```

```python
# pop() method to remove and return the last element
print(l)  # Print the original list
l.pop() # This will remove and return the last element of the list
```

```python
## removing javascript from the list
l.pop(4)  # Removes and returns the element at index 4
```

```python
print(l)
```

```python
print(l)
```

### Dictionary 
- Dictionary (the key-value pair):
- dict starts with curly brackets { }
- sytax ----  {key: value}
- used to store data with labels(keys) instead of positions or index
- Key must be strings/numbers/tuples and values can be anything
- keys  - does allow duplicates in the key values.
- Dictionary Methods 
  - .keys() -  returns all the keys in the dataset
  - .get() -  allow u to get the values using the keys as index
  - .values - returns all the values in the dataset

```python
empty_dict = {}  # Creating an empty dictionary
print(type(empty_dict))  # Checking the type of empty_dict
```

### Sets

- Set (it maintains the unique values)
- It has curly brackets but not key value pairs
-  to handle uniqueness
- removes duplicates automatically
- Note: it is Unordered
- it accepts only unique, immutable items
- Methods
  - add() - add a value to a set
  - sorted() - sort the values in a set

```python
guest = { "alfred", "emma", "oliver" , "ava", "james" , "alfred" }
print(guest)
print(type(guest))
```

```python
### how to add elements to a set
print(guest)
guest.add("sophia")
print(guest)

```

```python
sorted_guest = sorted(guest)
print(sorted_guest)
```

```python
### to reverse sort the set
reverse_sorted_guest = sorted(guest, reverse=True)  
print(reverse_sorted_guest)
print(type(reverse_sorted_guest))
```
