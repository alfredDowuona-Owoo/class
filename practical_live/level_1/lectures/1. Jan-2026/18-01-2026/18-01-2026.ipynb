#### advance control flow
##### if elif else

- if condition :
-    do something
- elif condition:
-    do something
- elif condition:
-     do something
- else:
-     do something


```python
operaton_name = "data"

if operaton_name == "addition":
    print("Performing addition operation")
elif operaton_name == "subtraction":
    print("Performing subtraction operation")
elif operaton_name == "divide":
    print("Performing division operation")
else:
    print("Unknown operation")
```

```python
operaton_name = "divide"


# CONTROL FLOW STARTS
if operaton_name == "addition":
    print("Performing addition operation")
elif operaton_name == "subtraction":
    print("Performing subtraction operation")
elif operaton_name == "divide":
    print("Performing division operation")
else:
    print("Unknown operation")
## CONTROL FLOW ENDS
print("program ended")
```
