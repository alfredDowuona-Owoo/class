```python
Is_the_AC_on = input("Enter (Yes/No) Is the AC on? : ")
Is_the_Coffee_maker_on = input("Enter (Yes/No) Is the Coffee Maker on? : ")
Is_the_Front_Door_unlocked = input("Enter (Yes/No) Is the Front Door unlocked?: ")
if Is_the_AC_on == "Yes":
    print("Turning off AC to save energy")
else:
    print("AC is off")
if Is_the_Coffee_maker_on == "Yes":
    print("Warning: Coffee Maker is still hot!")
else:
    print("Coffee Maker is off")
if Is_the_Front_Door_unlocked == "Yes":
    print("Locking the front door now.")
else:
    print("Front Door is locked")
```

```python
# get user input
extra_cheese = input("Extra cheese (True/False): ") == "True"
pepperoni = input("Pepperoni (True/False): ") == "True"
olives = input("Olives (True/False): ") == "True"

total_bill = 10

if extra_cheese:
    total_bill +=2

if pepperoni:
    total_bill +=3

if olives:
    total_bill+=1

print(total_bill)
```

```python
light_color = (input("Enter the traffic light color (red, yellow, Green): "))

if light_color == "Red":
    print("Full stop")
elif light_color == "Yellow":
    print("Slow down and prepare to stop.")
elif light_color == "Green":
    print("Proceed with caution.")
else:
    print("Error: Signal Malfunction!")
```

```python
Age = int(input("Please enter your age: "))

if Age < 5:
    print("Your ticket is free")
elif 5 <= Age <= 17 :
    print("Child Ticket: $10.")
elif 18 <= Age <= 60 :
    print("Adult Ticket: $18.")
else:
    print("Senior Ticket: $12.")
```
