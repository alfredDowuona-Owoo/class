### Scope of a variable
 - variables defined in a function is called a local variable of that function.
 - once the function ends the variable is forgotten

```python
## eg this gives us error because we are trying to assign to a variable before it is defined in the local scope

a=0

def add():
    a+=1
    print(a)
```

```python
add()
```

```python
### corrected code 
a=0  ## global variable

def add():
    a= 2 ##local to the function    
    a+=1
    print(a)

add()
```

```python
print(a)
add()
```

```python
a= [1,2,3,4,5]

def add():
    a.append(123)
    print(a)

print(a)
add()
print(a)
```

```python
a=0  ## global variable

## to use a global variable inside a function and modify it, we need to use the 'global' keyword
def add():
    global a ##local to the function    
    a+=1
    print(a)

print(a)
add()
print(a)
```

```python
a= []

def add():
    global a
    a.append(10)
    print(a)

print(a)
add()
print(a)
```

```python
def our_join_function(operating_system, *args):
    final_path=""
    for folder in args:
        if operating_system == "windows": 
            final_path = final_path + folder + "\\"
            if operating_system == "linux" or operating_system == "mac":
                final_path = final_path + folder + "/"
    final_path= final_path[:-1]  ## remove last slash
    return final_path
```

```python
f= open("dummy.txt")
while True:
    line = f.readline()
    # f.readline () for each line till end of file - 5th iteration
    # if condition is true it continues else breaks
    # if false it breaks
    if not line:
        break
    print(line.strip())  # strip() removes the newline character at the end of each line
f.close()
```
