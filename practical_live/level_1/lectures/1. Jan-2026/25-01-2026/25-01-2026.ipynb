### Control flow in loop

- Continue : - it skips are when it meets a condition
 - break : -  it breaks the loop itself / terminates the loop

```python
names = ["Alice", "Bob", "Charlie", "David", "eve", "Frank", "Grace", "Heidi", "Ivan", "Judy"]
```

```python
for name in names:
    if name == "eve":
        continue  # skip the name 'eve'
    name = name.capitalize()
    name = name.upper()
    print(name)
```

```python
for name in names:
    if name == "eve":
        break  # terminates when the name is 'eve'
    name = name.capitalize()
    name = name.upper()
    print(name)
```

```python
count = 0
print("start")
while True:
    if count > 100:
        break
    print("Count is:", count)
    count += 10
print("end")
```

```python
for i in range(1,11):
    print(i)
    for name in names:
        if name in ["eve", "Frank"]:
            break   # breaks the inner loop only
        print("  ", name)
```

```python
# Assignment

def add(a,b):
    print(a+b)

def subtract(a,b):
    print(a-b)

def divide(a,b):
    print(a//b)

def multiply(a,b):
    print(a*b)

def calculator():
    choice = int(input("Enter 1 for add, 2 for subtract, 3 for divide, 4 for multiple : "))

    value_1 = float(input("Enter first value : "))
    value_2 = float(input("Enter Second value : "))

    if choice == 1:
        add(value_1, value_2)
    elif choice == 2:
        subtract(value_1, value_2)
    elif choice == 3:
        divide(value_1, value_2)
    elif choice == 4:
        multiply(value_1, value_2)
    else:
        print("Please choose the correct option")
```

```python
calculator()
```
